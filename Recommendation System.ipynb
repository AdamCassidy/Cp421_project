```python
import pandas as pd

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_csv('ml-latest-small/ml-latest-small/ratings.csv', skiprows=1, sep=',', header=None, names=rnames)

ratings[:]
```

```python
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_csv('ml-latest-small/ml-latest-small/movies.csv', skiprows=1, sep=',', header=None, names=mnames,)

movies
```

##### Merge ratings and movies

```python
data = pd.merge(ratings, movies)
data
```

##### Random choose data for training and rest for test

```python
data_copy = data.copy()
train_set = data_copy.sample(frac=0.001) #random choose sample
test_set = data_copy.drop(train_set.index)

train_set[:]
```

##### This is the test set

```python
test_set[:]
```

##### Create pivot table for training set and convert it to numpy array

```python
train_table = pd.pivot_table(train_set, values='rating', index='user_id', columns='title')
train_table
```

```python
movies = []
for col in train_table.columns:
    movies.append(col)
    #print(col)
#print(movieId)
#print(len(movieId))

train_array = train_table.to_numpy()
print(train_array)
print(len(train_array))
print(len(train_array[0]))

#print(train_array.dtype)
```

##### The function for covert rating data to binary data.
##### If the rate above the user's average rate equal to 1, else set to 'nan' (we only need positive rate)

```python
import numpy as np
def ratingToBinary(array):
    userNo = len(array)
    mean = np.nanmean(array, axis = 1)
    #print(mean)
    for user in range(0, userNo):
        movieNo = len(array[user])
        for movie in range(0, movieNo):
            if not np.isnan(array[user][movie]):
                if array[user][movie] >= mean[user]:
                    array[user][movie] = 1
                else:
                    array[user][movie] = np.nan
     #   print(array[user])
                    
#ratingToBinary(train_array)
```

```python
print(train_array)
ratingToBinary(train_array)
print(train_array)
```

##### Replace 1 with movie name

```python
train = []
for user in range(0, len(train_array)):
    u = []
    for movie in range(0, len(train_array[user])):
        if train_array[user][movie] == 1.0:
            u.append(movies[movie])
        else:
            u.append(str(train_array[user][movie]))
    train.append(u)
    
```

#### Using apriori algorithm get rules

```python
from apyori import apriori
minSup = 0.001
minConf = 0.2
minLift = 1.1
minLen = 2

rules = apriori(train, min_support = minSup, min_confidence = minConf, min_lift = minLift, min_length = minLen)
print(rules)
res = list(rules)
print(len(res))
print(res[0])
```
